```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv("loan_approval_dataset.csv")
df.head()
#df[' loan_status']
```

```python
df.drop(['loan_id'],axis=1,inplace=True)
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[' loan_status'] = df[' loan_status'].map({" Approved": 1, " Rejected": 0})
df[" education"] = df[" education"].map({" Graduate": 1, " Not Graduate": 0})
df[" self_employed"] = df[" self_employed"].map({" Yes": 1, " No": 0})
print(df.dtypes)
```

```python
#df.columns
#df[' loan_status']
df[' self_employed']
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x=" loan_status", data=df)
plt.title("Loan Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation=df.corr()
sns.heatmap(correlation, annot=True,linewidths=0.01)
plt.title("Feature Correlation Heatmap")
plt.show()
print(correlation[' loan_status'].sort_values(ascending=False))
```

```python
columns_to_drop = [
    ' luxury_assets_value',   # High correlation with income_annum and loan_amount
    ' income_annum',          # High correlation with loan_amount and luxury_assets_value
    ' bank_asset_value'       # Low correlation with loan_status
]
df = df.drop(columns=columns_to_drop)
```
